```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df=pd.read_csv('/jobs_in_data.csv')
fig=plt.figure(figsize=(8,5))
g=sns.heatmap(df.corr());
g.set_title('correlation between continuous variables')
```

```python
import matplotlib.pyplot as plt
fig,axis =plt.subplots(1,2,figsize=(10,10))
ordered_data=df['work_year'].value_counts()
g=sns.countplot(y='work_year',data=df,ax=axis[0],order=ordered_data.index)
axis[0].set_title("The frequency of each year")

avg_salary_by_year=df.groupby('work_year')['salary'].mean().reset_index()
axis[1].plot(avg_salary_by_year['work_year'],avg_salary_by_year['salary'],marker='o')
axis[1].set_title("Avarage salary over year")
axis[1].set_xlabel('work year')
axis[1].set_ylabel('avarage salary')
plt.show()
```

```python
D=df[df['work_year']==2023]
g=sns.boxplot(y='salary',data=D)
g.set_title( "Salary distribution for 2023")
```

```python
fig=plt.figure(figsize=(15,15))
freq_of_jobs=df['job_title'].value_counts().head(50)

plt.barh(freq_of_jobs.index, freq_of_jobs)

plt.title('Job Title Distribution for top 50 jobs')
plt.xlabel('Job Title')
plt.ylabel('Frequency')

plt.show()
```

```python
avg_salaries=df.groupby('job_title')['salary'].mean().sort_values(ascending=False)
top10=avg_salaries.head(10)
bottom10=avg_salaries.tail(10)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))


sns.barplot(x=top10.index, y=top10.values, ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_title('Top 10 Job Titles with Highest Avg Salaries')
axes[0].set_xlabel('Job Title')
axes[0].set_ylabel('Avarage salary')

sns.barplot(x=bottom10.index,y=bottom10.values,ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_title('Bottom 10 Job Titles with Least Avg Salaries')
axes[1].set_xlabel('Job title')
axes[1].set_ylabel('Avarage salary')

plt.tight_layout()
plt.show()
```

```python
Ex=df[df['job_title']=='Data Engineer']
experience_counts=Ex['experience_level'].value_counts().reset_index()
experience_counts.columns=['experience','frequency']
G=px.pie(experience_counts,values='frequency',names='experience',title='Experience Level Distribution for Data Engineers')
G.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
df=pd.read_csv('/jobs_in_data.csv')
Ex=df[df['job_title']=='Data Engineer']
fig=plt.figure(figsize=(10,8))
G=sns.catplot(x='work_setting',data=Ex,kind='count')
G.fig.suptitle('work_setting Distribution for Data Engineers')
```

```python
fig=plt.figure(figsize=(10,8))
G=sns.catplot(x='work_setting',data=df,kind='count')
G.fig.suptitle('work_setting Distribution')
```

```python
avg_by_currency=df.groupby('salary_currency')['salary_in_usd'].mean().reset_index()
avg_by_currency=avg_by_currency.sort_values(by='salary_in_usd',ascending=True)
g=plt.barh(y=avg_by_currency['salary_currency'],width=avg_by_currency['salary_in_usd'])
plt.title=('Avarage salaries in usd by salary currencies vs currency')
plt.xlabel('salary in usd')
plt.ylabel('salary currency')
plt.show()
```
